````python
# 3.Problemin Çözümünü Buraya Yazınız
class Araç:
    def __init__(self, İsim, Hız, Kilometre):
        self.İsim = İsim
        self.Hız = Hız
        self.Kilometre = Kilometre
        
    def display_info(self):
        print("İsim:", self.İsim, "Hız:", self.Hız, "Kilometre:", self.Kilometre)

class Otobüs(Araç):
    pass

bus = Otobüs("School Volvo", 180, 12)
bus.display_info()



````
