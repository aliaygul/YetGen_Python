# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time


class Kumanda:

    
    def __init__(self):
        self.tv_durumu= "Kapalı"
        self.ses=5
        self.kanal_listesi=["CNN","FOX","HALK TV","ATV","SHOW"]
        self.kanal="TRT"
    
    def tv_ac(self):
        if not self.tv_durumu: 
            print("Televizyon açılıyor...")
            self.tv_durumu == True
            
        else:
            print("Televizyon zaten açık...")
            
    
    def tv_kapali(self):
        if self.tv_durumu:
            print("Televizyon kapanıyor...")
            self.tv_durumu == False
        else:
            print("Televizyon zaten kapalı...")
    
    def ses_ayari(self):
        while True:
            cevap = input("Sesi azalt: '<'\nSesi arttır: '>'\nÇıkış yapamk için 'q' tuşuna basın\n")
            if cevap =="<":
                if self.ses !=0:
                    self.ses-=1
                    print("Ses:", self.ses)
                elif self.ses==0:
                    print("Ses zaten en düşük seviyede!")
            elif cevap == ">":
                if self.ses !=100:
                    self.ses +=1
                    print("Ses:", self.ses)
            else:
                print("Ses Güncellendi:", self.ses)
                break
    
    def kanal_ekle(self,kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print("Kanal eklendi.")
    
    def kanal_sayisi(self):
        print("Kanal Sayısı:", len(self.kanal_listesi))
    
    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1) #parantez içerisine yazılan min ve max değerleri arasında rastgele tamsayı (integer) tipinde sayı üretir
        self.kanal= self.kanal_listesi[rastgele]
        print("Şu anki kanal:", self.kanal)

    def tv_bilgileri(self):
        print("TV Durumu:", self.tv_durumu)
        print("Ses Seviyesi:", self.ses)
        print("Kanal Listesi:", self.kanal_listesi)
        print("Şu anki kanal:", self.kanal)

kumanda=Kumanda()

print("Televizyon Kumandasına Hoşgeldiniz.")

while True:
    print("Tv açmak için 'a'\nTv kapatmak için 'k'\nSes ayarları için 's'\nKanal eklemek için 'e'\nKanal sayısını öğrenmek için 'c'\nRastgele kanala geçmek için 'r'\nTV bilgileri için'i'\nÇıkış yapmak için 'q' tuşuna basın\n")
    işlem=input("Yapmak istediğiniz işlemi giriniz:")

    if işlem == "q":
        print("Çıkış yapılıyor")
        time.sleep(1)
        break
    
    elif işlem == "a":
        print(kumanda.tv_ac())
    
    elif işlem == "k":
        kumanda.tv_kapali()

    elif işlem == "s":
        kumanda.ses_ayari()

    elif işlem =="e":
        kanal_isimleri= input("Eklemek istediğiniz kanalları ',' iler ayırarak giriniz:")
        kanal_listesi=kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            kumanda.kanal_ekle(kanal)
    
    elif işlem == "c":
        kumanda.kanal_sayisi()
    
    elif işlem == "r":
        kumanda.rastgele_kanal()

    elif işlem == "i":
        kumanda.tv_bilgileri()
```
